```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
path = 'data_for_ariam/'
tickers = ['AAPL', 'KO', 'NVDA', 'PFE', 'TSLA']
# 第一步：读取数据
all_df = {}
for i in range(5):
    df = pd.read_csv(path+tickers[i]+'.csv', encoding='utf-8', index_col=0, parse_dates=True)
    print(df.head(5))
    all_df[tickers[i]]=df
all_df
```

```python
models = ['LR', 'ARIAM', 'LSTM', 'LR_only_close', 'ARIAM_2_2', 'LSTM_L_64', 'LSTM_L_256', 'LSTM_step_20', 'LSTM_step_25']
all_predict_result = {}
for model in models:
    df = pd.read_csv(model+'_result.csv', encoding='utf-8')
    df.head(5)
    all_predict_result[model] = df
all_predict_result
```

```python
tmp = all_predict_result['LSTM']['date'].drop_duplicates()
dates = list(tmp)
type(dates[0])
len(all_predict_result['LSTM'])
dates
```

```python
result_for_compare = {}
for model in models:
    result_for_compare[model] = {}
    for stock in tickers:
        result_for_compare[model][stock] = []
    print(model, len(all_predict_result[model]))
    for i in range(len(all_predict_result[model])):
        row = all_predict_result[model].iloc[i]
        if row['date'] in dates:
            result_for_compare[model][row['stock_name']].append(float(row['predict']))
    for stock in tickers:
        print(stock, len(result_for_compare[model][stock]))
result_for_compare
```

```python
all_data = {}
for i in range(5):
    all_data[tickers[i]] = list(all_df[tickers[i]]['Close'])[-30:]
    #print(len(all_data[tickers[i]]))
    #print(all_data[tickers[i]][0]-all_data[tickers[i]][1])
all_data
```

```python
all_diff = {}
mse = {}
for stock in tickers:
    all_diff[stock] = {}
    mse[stock] = {}
    for model in models:
        all_diff[stock][model] = []
        mse[stock][model] = 0.0
        for i in range(30):
            diff = all_data[stock][i] - result_for_compare[model][stock][i]
            all_diff[stock][model].append(diff)
            mse[stock][model] += diff*diff
        mse[stock][model] /= 30
print(mse)
print(all_diff)
```

```python
def show_result(truths, preds, days, stock_sym, figname):
    plt.figure(figsize=(12, 6))
    plt.plot(days, truths, label='truth', linewidth=1)#, linestyle='--')
    plt.plot(days, preds, label='pred', linewidth=1)#, linestyle=':')
    plt.legend(loc='upper left', frameon=False)
    plt.xlabel("day")
    plt.ylabel("price")
    plt.ylim((min(truths), max(truths)))
    plt.grid(ls='--')
    
    if stock_sym:
        plt.title(stock_sym + " | Last %d days in test" % len(truths))
    #plt.show()
    plt.savefig(figname, format='png', bbox_inches='tight', transparent=True)
    plt.close()
days = range(30)
```

```python
days = range(30)
for stock in tickers:
    show_result(all_data[stock], result_for_compare['ARIAM'][stock], days, stock, 'evaluation/ARIAM/'+stock+'.png')
```
